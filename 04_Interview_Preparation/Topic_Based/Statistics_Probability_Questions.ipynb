## Question 1: Expected Value of a Trading Strategy

**Question:** Your trading strategy has a 60% win rate. When you win, you make $100. When you lose, you lose $80. What is the expected value per trade? Should you trade this strategy?

```python
# Question 1: Expected Value Calculation
print("\n" + "="*60)
print("QUESTION 1: Expected Value of Trading Strategy")
print("="*60)

win_rate = 0.60
win_amount = 100
loss_amount = -80

expected_value = win_rate * win_amount + (1 - win_rate) * loss_amount

print(f"\nGiven:")
print(f"  Win rate: {win_rate*100}%")
print(f"  Win amount: ${win_amount}")
print(f"  Loss amount: ${abs(loss_amount)}")

print(f"\nSolution:")
print(f"  E[X] = P(win) Ã— Win + P(loss) Ã— Loss")
print(f"  E[X] = {win_rate} Ã— ${win_amount} + {1-win_rate} Ã— (${loss_amount})")
print(f"  E[X] = ${win_rate * win_amount} + ${(1-win_rate) * loss_amount}")
print(f"  E[X] = ${expected_value:.2f}")

print(f"\nâœ… Answer: Expected value = ${expected_value:.2f} per trade")
print(f"   {'Yes, trade this strategy!' if expected_value > 0 else 'Do NOT trade this strategy'}")
```

## Question 2: Coin Flip Game

**Question (Jane Street):** You flip a fair coin. If it's heads, you win $1. If it's tails, you flip again. On the second flip, heads wins $2, tails flips again. This continues, doubling the prize. What's the expected value of this game?

```python
# Question 2: Coin Flip Game (St. Petersburg Paradox variant)
print("\n" + "="*60)
print("QUESTION 2: Coin Flip Game (Jane Street Style)")
print("="*60)

print("\nAnalysis:")
print("  E[X] = (1/2)Ã—$1 + (1/4)Ã—$2 + (1/8)Ã—$4 + ...")
print("  E[X] = Î£ (1/2^n) Ã— 2^(n-1) for n=1 to âˆž")
print("  E[X] = Î£ (1/2) for n=1 to âˆž")
print("  E[X] = 1/2 + 1/2 + 1/2 + ... = âˆž")

# Simulate to show convergence is slow
def simulate_game(n_simulations=10000, max_flips=30):
    payoffs = []
    for _ in range(n_simulations):
        prize = 1
        for _ in range(max_flips):
            if np.random.random() < 0.5:  # Heads
                payoffs.append(prize)
                break
            prize *= 2
        else:
            payoffs.append(prize)  # Hit max flips
    return payoffs

payoffs = simulate_game()
print(f"\nSimulation (10,000 games, max 30 flips):")
print(f"  Mean payoff: ${np.mean(payoffs):.2f}")
print(f"  Median payoff: ${np.median(payoffs):.2f}")
print(f"  Max payoff: ${np.max(payoffs):.2f}")

print(f"\nâœ… Answer: Theoretically infinite, but practically bounded")
print(f"   Interview insight: Discuss how risk-averse players would pay finite amount")
```

## Question 3: Hypothesis Testing for Trading

**Question:** Your trading strategy generated a 15% return over 252 trading days with daily volatility of 1.5%. Is this statistically significant at the 5% level?

```python
# Question 3: Statistical Significance of Returns
print("\n" + "="*60)
print("QUESTION 3: Is the Return Statistically Significant?")
print("="*60)

annual_return = 0.15  # 15%
n_days = 252
daily_vol = 0.015  # 1.5%

# Calculate daily return
daily_return = annual_return / n_days

# T-statistic under null hypothesis (mean return = 0)
t_stat = daily_return / (daily_vol / np.sqrt(n_days))

# P-value (two-tailed)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=n_days-1))

print(f"\nGiven:")
print(f"  Annual return: {annual_return*100}%")
print(f"  Trading days: {n_days}")
print(f"  Daily volatility: {daily_vol*100}%")

print(f"\nSolution:")
print(f"  Daily return = {annual_return*100}% / {n_days} = {daily_return*100:.4f}%")
print(f"  Standard error = Ïƒ / âˆšn = {daily_vol} / âˆš{n_days} = {daily_vol/np.sqrt(n_days):.6f}")
print(f"  T-statistic = {daily_return:.6f} / {daily_vol/np.sqrt(n_days):.6f} = {t_stat:.2f}")
print(f"  P-value = {p_value:.4f}")

print(f"\nâœ… Answer: T-stat = {t_stat:.2f}, p-value = {p_value:.4f}")
print(f"   {'Significant at 5% level!' if p_value < 0.05 else 'NOT significant at 5% level'}")
```

## Question 4: Sharpe Ratio Confidence Interval

**Question:** You calculated a Sharpe ratio of 1.5 based on 3 years of daily data. What's the 95% confidence interval?

```python
# Question 4: Sharpe Ratio Confidence Interval
print("\n" + "="*60)
print("QUESTION 4: Sharpe Ratio Confidence Interval")
print("="*60)

sharpe = 1.5
n_years = 3
n_obs = n_years * 252

# Standard error of Sharpe ratio (approximation)
# SE(SR) â‰ˆ sqrt((1 + SRÂ²/2) / n)
se_sharpe = np.sqrt((1 + sharpe**2 / 2) / n_obs)

# 95% CI
z_95 = 1.96
ci_lower = sharpe - z_95 * se_sharpe
ci_upper = sharpe + z_95 * se_sharpe

print(f"\nGiven:")
print(f"  Sharpe ratio: {sharpe}")
print(f"  Years of data: {n_years} ({n_obs} observations)")

print(f"\nSolution:")
print(f"  SE(SR) â‰ˆ âˆš[(1 + SRÂ²/2) / n]")
print(f"  SE(SR) = âˆš[(1 + {sharpe}Â²/2) / {n_obs}]")
print(f"  SE(SR) = {se_sharpe:.4f}")
print(f"  95% CI = SR Â± 1.96 Ã— SE")
print(f"  95% CI = {sharpe} Â± {z_95 * se_sharpe:.4f}")

print(f"\nâœ… Answer: 95% CI = [{ci_lower:.2f}, {ci_upper:.2f}]")
print(f"   Key insight: Sharpe ratios are highly uncertain!")
```

## Question 5: Kelly Criterion

**Question:** You have a strategy with 55% win rate, winning $1 and losing $1. What fraction of your bankroll should you bet according to Kelly?

```python
# Question 5: Kelly Criterion
print("\n" + "="*60)
print("QUESTION 5: Kelly Criterion Optimal Bet Size")
print("="*60)

p = 0.55  # Win probability
b = 1  # Win/loss ratio (win $1, lose $1)

# Kelly formula: f* = (bp - q) / b = p - q = 2p - 1 (when b=1)
kelly_fraction = (b * p - (1 - p)) / b

print(f"\nGiven:")
print(f"  Win probability (p): {p*100}%")
print(f"  Win/Loss ratio (b): {b}")

print(f"\nKelly Formula:")
print(f"  f* = (bÃ—p - q) / b")
print(f"  f* = ({b}Ã—{p} - {1-p}) / {b}")
print(f"  f* = ({b*p} - {1-p}) / {b}")
print(f"  f* = {kelly_fraction:.2f}")

print(f"\nâœ… Answer: Optimal bet size = {kelly_fraction*100:.0f}% of bankroll")
print(f"   Practical note: Most traders use half-Kelly (={kelly_fraction/2*100:.0f}%) for safety")
```

## Question 6: Correlation vs Causation

**Question (Two Sigma):** You find that ice cream sales are correlated with drowning deaths. What's wrong with concluding that ice cream causes drowning?

```python
# Question 6: Spurious Correlation Demo
print("\n" + "="*60)
print("QUESTION 6: Correlation vs Causation")
print("="*60)

# Simulate the confounding variable scenario
np.random.seed(42)
n = 365

# Temperature is the confounding variable
temperature = 50 + 30 * np.sin(np.linspace(0, 2*np.pi, n)) + np.random.normal(0, 5, n)

# Ice cream sales depend on temperature
ice_cream = 100 + 2 * temperature + np.random.normal(0, 20, n)

# Drowning deaths depend on temperature (more swimming)
drowning = 5 + 0.1 * temperature + np.random.normal(0, 2, n)

# Calculate correlations
corr_ice_drowning = np.corrcoef(ice_cream, drowning)[0, 1]
corr_temp_ice = np.corrcoef(temperature, ice_cream)[0, 1]
corr_temp_drowning = np.corrcoef(temperature, drowning)[0, 1]

print(f"\nCorrelation Analysis:")
print(f"  Ice Cream â†” Drowning:  r = {corr_ice_drowning:.3f} (high!)")
print(f"  Temperature â†” Ice Cream: r = {corr_temp_ice:.3f}")
print(f"  Temperature â†” Drowning:  r = {corr_temp_drowning:.3f}")

print(f"\nâœ… Answer: Temperature is a CONFOUNDING VARIABLE")
print(f"   - Both ice cream sales AND drowning increase with temperature")
print(f"   - No causal relationship between ice cream and drowning")
print(f"   - In trading: Beware spurious correlations in backtests!")
```
