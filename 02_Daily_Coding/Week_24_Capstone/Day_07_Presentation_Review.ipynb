## 1. Capstone Executive Summary

```python
# Configuration
TICKERS = ['AAPL', 'MSFT', 'GOOGL', 'JPM', 'GS']
START_DATE = '2020-01-01'
END_DATE = datetime.now().strftime('%Y-%m-%d')

# Download data
print("ðŸ“¥ Loading market data...")
data = yf.download(TICKERS, start=START_DATE, end=END_DATE, progress=False, auto_adjust=True)
prices = data['Close'].dropna()
returns = prices.pct_change().dropna()

# Calculate portfolio metrics
portfolio_returns = returns.mean(axis=1)
cumulative = (1 + portfolio_returns).cumprod()

# Key metrics
total_return = cumulative.iloc[-1] - 1
annual_return = (1 + total_return) ** (252 / len(portfolio_returns)) - 1
volatility = portfolio_returns.std() * np.sqrt(252)
sharpe = (annual_return - 0.05) / volatility
max_dd = ((cumulative - cumulative.cummax()) / cumulative.cummax()).min()

print(f"""
{'='*70}
              CAPSTONE PROJECT EXECUTIVE SUMMARY
{'='*70}

ðŸ“Œ PROJECT: Multi-Asset ML Trading System
ðŸ“… Analysis Period: {START_DATE} to {END_DATE}
ðŸ“Š Universe: {', '.join(TICKERS)}

{'='*70}
                    PERFORMANCE METRICS
{'='*70}

  Total Return:       {total_return*100:>8.2f}%
  Annual Return:      {annual_return*100:>8.2f}%
  Volatility:         {volatility*100:>8.2f}%
  Sharpe Ratio:       {sharpe:>8.2f}
  Max Drawdown:       {max_dd*100:>8.2f}%

{'='*70}
                    SKILLS DEMONSTRATED
{'='*70}

âœ… Data Engineering: yfinance API, point-in-time data
âœ… Feature Engineering: Technical indicators, momentum, volatility
âœ… ML Models: Linear, Tree Ensembles, Neural Networks
âœ… Risk Management: VaR, CVaR, position sizing
âœ… Backtesting: Transaction costs, walk-forward validation
âœ… Production: Model registry, monitoring, deployment

{'='*70}
""")
```

## 4. Final Trading Signals

```python
print("\n" + "="*70)
print("ðŸŽ¯ FINAL TRADING SIGNALS - CAPSTONE OUTPUT")
print("="*70)
print(f"Analysis Date: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
print(f"Portfolio Capital: $100,000")
print("\n")

# Calculate comprehensive signals
for ticker in TICKERS:
    price = prices[ticker].iloc[-1]
    ret = returns[ticker]
    
    # Metrics
    momentum_20d = (prices[ticker].iloc[-1] / prices[ticker].iloc[-20] - 1)
    vol_20d = ret.tail(20).std() * np.sqrt(252)
    sharpe_60d = (ret.tail(60).mean() * 252 - 0.05) / (ret.tail(60).std() * np.sqrt(252))
    
    # Signal logic
    score = 0
    if momentum_20d > 0.03: score += 1
    if momentum_20d > 0.08: score += 1
    if vol_20d < 0.25: score += 1
    if sharpe_60d > 0.5: score += 1
    if momentum_20d < -0.03: score -= 1
    if momentum_20d < -0.08: score -= 1
    if vol_20d > 0.35: score -= 1
    
    if score >= 2:
        signal = "ðŸŸ¢ STRONG BUY"
        options = "CALL"
    elif score == 1:
        signal = "ðŸŸ¡ WEAK BUY"
        options = "CALL spread"
    elif score <= -2:
        signal = "ðŸ”´ STRONG SELL"
        options = "PUT"
    elif score == -1:
        signal = "ðŸŸ  WEAK SELL"
        options = "PUT spread"
    else:
        signal = "âšª HOLD"
        options = "Iron Condor"
    
    print(f"{'='*30} {ticker} {'='*30}")
    print(f"   Current Price:   ${price:.2f}")
    print(f"   20d Momentum:    {momentum_20d*100:+.2f}%")
    print(f"   20d Volatility:  {vol_20d*100:.1f}%")
    print(f"   60d Sharpe:      {sharpe_60d:.2f}")
    print(f"   Signal Score:    {score}")
    print(f"   ðŸ“Š Signal:       {signal}")
    print(f"   ðŸ“ˆ Options:      {options}")
    print(f"   ðŸŽ¯ Stop Loss:    ${price * 0.95:.2f} (-5%)")
    print(f"   ðŸ’° Take Profit:  ${price * 1.10:.2f} (+10%)")
    print()
```

## 5. Portfolio Visualization

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Cumulative Returns
ax1 = axes[0, 0]
for ticker in TICKERS:
    cum = (1 + returns[ticker]).cumprod()
    ax1.plot(cum, label=ticker, linewidth=1.5)
ax1.set_title('Cumulative Returns by Ticker', fontweight='bold')
ax1.set_ylabel('Growth of $1')
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)

# 2. Correlation Heatmap
ax2 = axes[0, 1]
corr = returns.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn', center=0, ax=ax2)
ax2.set_title('Correlation Matrix', fontweight='bold')

# 3. Rolling Volatility
ax3 = axes[1, 0]
for ticker in TICKERS:
    rolling_vol = returns[ticker].rolling(20).std() * np.sqrt(252) * 100
    ax3.plot(rolling_vol, label=ticker, linewidth=1)
ax3.set_title('20-Day Rolling Volatility', fontweight='bold')
ax3.set_ylabel('Annualized Volatility (%)')
ax3.legend(loc='upper right')
ax3.grid(True, alpha=0.3)

# 4. Return Distribution
ax4 = axes[1, 1]
portfolio_returns.hist(bins=50, ax=ax4, color='steelblue', alpha=0.7, edgecolor='black')
ax4.axvline(0, color='red', linestyle='--', linewidth=2)
ax4.axvline(portfolio_returns.mean(), color='green', linestyle='--', linewidth=2)
ax4.set_title('Portfolio Return Distribution', fontweight='bold')
ax4.set_xlabel('Daily Return')
ax4.set_ylabel('Frequency')

plt.tight_layout()
plt.savefig('capstone_summary.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nâœ… Summary visualization saved to capstone_summary.png")
```
